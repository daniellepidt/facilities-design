# Plot tests file:
This file is used to try out various ways of plotting the simulation on the fly.

```python
import numpy as np
c = np.random.randint(90, 120, size=(8,2,16))
c[c < 100] = 0
c
```

```python
flat = c.flatten()
flat = flat[flat != 0]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(projection='3d')

z,x,y = c.nonzero() 
scatter = ax.scatter(x,y,z, c=flat, alpha=1)

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper center", title="Item Type", ncol=5, fancybox=True, shadow=True)
ax.add_artist(legend1)

ax.set_xlabel('Aisle Side')
ax.set_ylabel('Aisle Depth')
ax.set_zlabel('Aisle Height')
plt.xticks([0,1], ['Left', 'Right'])
plt.yticks(np.arange(1,17, step=2), np.arange(1,17, step=2))
ax.set_zticks(np.arange(0,8), np.arange(1,9))
plt.show()
```
